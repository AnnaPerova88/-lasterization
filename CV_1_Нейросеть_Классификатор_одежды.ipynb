```python
# ============================================
# БЛОК 1: ИМПОРТ БИБЛИОТЕК
# ============================================
# TensorFlow/Keras - библиотека для создания нейронных сетей
# Это как Lego для ML: собираем модель из готовых блоков
import tensorflow as tf
from tensorflow import keras  # Keras - упрощённый интерфейс для TensorFlow

# NumPy - работа с числами и матрицами (изображения = матрицы пикселей)
import numpy as np

# Matplotlib - рисование графиков и показ картинок
import matplotlib.pyplot as plt

# ============================================
# БЛОК 2: ЗАГРУЗКА ДАННЫХ
# ============================================
# Fashion-MNIST - датасет из 70,000 фотографий одежды
# 10 категорий: футболки, брюки, платья, обувь и т.д.
# Каждая картинка 28x28 пикселей (чёрно-белая)

# Загружаем данные: 60,000 для обучения, 10,000 для теста
fashion_mnist = keras.datasets.fashion_mnist
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Что такое X и y?
# X_train - картинки одежды (features/признаки)
# y_train - правильные ответы: какая это одежда (labels/метки)

# Названия категорий для понимания
class_names = ['Футболка', 'Брюки', 'Свитер', 'Платье', 'Пальто',
               'Сандалии', 'Рубашка', 'Кроссовки', 'Сумка', 'Ботинки']

# ============================================
# БЛОК 3: РАЗВЕДКА ДАННЫХ (EDA)
# ============================================
# Посмотрим на первые 9 картинок
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.imshow(X_train[i], cmap='gray')  # cmap='gray' - чёрно-белое отображение
    plt.title(class_names[y_train[i]])   # Показываем правильное название
    plt.axis('off')  # Убираем оси координат
plt.tight_layout()
plt.show()

# Проверяем размеры данных
print(f"Размер обучающей выборки: {X_train.shape}")
# (60000, 28, 28) = 60,000 картинок по 28x28 пикселей
print(f"Размер тестовой выборки: {X_test.shape}")

# ============================================
# БЛОК 4: ПОДГОТОВКА ДАННЫХ
# ============================================
# Нормализация: переводим значения пикселей из диапазона 0-255 в 0-1
# Зачем? Нейросети легче учиться на маленьких числах
X_train = X_train / 255.0
X_test = X_test / 255.0

# ============================================
# БЛОК 5: СОЗДАНИЕ МОДЕЛИ
# ============================================
# Sequential - модель, где слои идут друг за другом (как конвейер)
model = keras.Sequential([
    # Flatten - "расплющивает" картинку 28x28 в вектор из 784 чисел
    # Нейросеть не понимает 2D, ей нужен 1D вектор
    keras.layers.Flatten(input_shape=(28, 28)),

    # Dense - полносвязный слой (каждый нейрон связан со всеми)
    # 128 нейронов, activation='relu' - функция активации (пропускает только положительные)
    keras.layers.Dense(128, activation='relu'),

    # Dropout - случайно "выключает" 20% нейронов во время обучения
    # Это предотвращает переобучение (когда модель запоминает, а не учится)
    keras.layers.Dropout(0.2),

    # Выходной слой: 10 нейронов (по одному на каждую категорию одежды)
    # softmax - превращает выходы в вероятности (сумма = 1)
    keras.layers.Dense(10, activation='softmax')
])

# ============================================
# БЛОК 6: КОМПИЛЯЦИЯ МОДЕЛИ
# ============================================
# optimizer='adam' - алгоритм обучения (самый популярный, работает хорошо)
# loss='sparse_categorical_crossentropy' - функция потерь для многоклассовой классификации
# metrics=['accuracy'] - метрика качества (точность в %)
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# ============================================
# БЛОК 7: ОБУЧЕНИЕ
# ============================================
# epochs=10 - проходим весь датасет 10 раз
# validation_data - проверяем качество на тестовых данных после каждой эпохи
history = model.fit(X_train, y_train,
                    epochs=10,
                    validation_data=(X_test, y_test))

# ============================================
# БЛОК 8: ОЦЕНКА КАЧЕСТВА
# ============================================
# Тестируем модель
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f'\nТочность на тестовых данных: {test_acc*100:.2f}%')

# ============================================
# БЛОК 9: ВИЗУАЛИЗАЦИЯ ОБУЧЕНИЯ
# ============================================
# Рисуем график точности по эпохам
plt.figure(figsize=(12, 4))

# График точности
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Обучение')
plt.plot(history.history['val_accuracy'], label='Валидация')
plt.title('Точность модели')
plt.xlabel('Эпоха')
plt.ylabel('Accuracy')
plt.legend()

# График потерь
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Обучение')
plt.plot(history.history['val_loss'], label='Валидация')
plt.title('Потери модели')
plt.xlabel('Эпоха')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# ============================================
# БЛОК 10: ПРЕДСКАЗАНИЯ НА НОВЫХ ДАННЫХ
# ============================================
# Делаем предсказания для первых 10 картинок из теста
predictions = model.predict(X_test[:10])

# Показываем результаты
plt.figure(figsize=(15, 10))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(X_test[i], cmap='gray')

    # Находим индекс максимальной вероятности
    predicted_label = np.argmax(predictions[i])
    true_label = y_test[i]

    # Цвет заголовка: зелёный - правильно, красный - ошибка
    color = 'green' if predicted_label == true_label else 'red'
    plt.title(f"Пред: {class_names[predicted_label]}\nИстин: {class_names[true_label]}",
              color=color)
    plt.axis('off')

plt.suptitle('Результаты классификации (зелёный = верно, красный = ошибка)',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# ============================================
# БЛОК 11: СОХРАНЕНИЕ МОДЕЛИ ДЛЯ ПОРТФОЛИО
# ============================================
# Сохраняем модель в файл .h5
model.save('fashion_classifier.h5')
print("✅ Модель сохранена! Теперь её можно загрузить и использовать")

# Для портфолио: создаём README
print("\n📋 ДЛЯ ПОРТФОЛИО:")
print("Проект: Классификатор одежды Fashion-MNIST")
print("Технологии: TensorFlow, Keras, Python")
print("Точность: {:.2f}%".format(test_acc*100))
print("Описание: Нейронная сеть для автоматической категоризации товаров одежды")
print("Применение: E-commerce, автоматическая разметка каталогов")
```
